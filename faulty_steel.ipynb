```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('data/faults.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
fault_cols = ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']
```

```python
df['Fault_Types'] = df[fault_cols].idxmax(axis=1)
```

```python
df.sample(5)
```

```python
df.drop(columns = fault_cols, inplace=True)
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df['Fault_Types'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Fault_Types'] = le.fit_transform(df['Fault_Types'])
```

```python
df.describe()
```

```python
X = df.drop(columns=['Fault_Types'])
y = df['Fault_Types']
```

```python
binary_cols = ['TypeOfSteel_A300', 'TypeOfSteel_A400']
num_cols = X.columns.difference(binary_cols)
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), num_cols)
    ],
    remainder='passthrough'   # keeps binary columns as-is
)

X_scaled = preprocessor.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(
    X_scaled,
    columns=list(num_cols) + list(binary_cols)
)
```

```python
X_scaled.head()
y.sample(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf1 = RandomForestClassifier()

rf1.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report 

y_pred = rf1.predict(X_test)

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}
```

```python
grid = GridSearchCV(
    estimator=rf1,
    param_grid=param_grid,
    cv=5,
    scoring='f1_macro',
    n_jobs=-1,
    verbose=2
)

grid.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_test_pred = best_model.predict(X_test)

print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
```

```python
# baseline model
rf_baseline = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_baseline.fit(X_train, y_train)

y_base_pred = rf_baseline.predict(X_test)

print("Baseline:")
print(classification_report(y_test, y_base_pred))

print("\nTuned:")
print(classification_report(y_test, y_test_pred))
```
